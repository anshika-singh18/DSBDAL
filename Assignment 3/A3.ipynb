Manual Calculation of the statitics

```python
def Mean(array:np.ndarray):
    length = len(array)
    s=0
    for val in array:
        s+=val
    result = f"Mean is{s/length}"
    return result

def Median(array:np.ndarray):
    copy = array.copy()
    copy.sort()
    length = len(array)
    if length%2==0:
        result = f"Median is {(copy[length//2] + copy[(length//2)-1])}"
    else:
        result = f"Median is {copy[length//2]}"
    return result
def Mode(array:np.ndarray):
    modeMap = dict()
    for val in array:
        if val in modeMap:
            modeMap[val]+=1
        else:
            modeMap[val] = 1
    maxCount = max(modeMap.values())
    mode = [key for key, values in modeMap.items() if value==maxCount]
    result = f"Mode value/s is/are: {mode}"
    return result
def myStdDev(array:np.ndarray):
    length = len(array)
    s=0
    for val in array:
        s+=val
    mean = s/length
    deviations = [(x-mean)**2 for x in array]
    variance = sum(deviations) / length
    stdDeviation = variance**0.5
    result = f"Std. Deviation is {stdDeviation}"
    return result
```
